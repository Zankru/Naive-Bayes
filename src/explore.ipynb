```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/naive-bayes-project-tutorial/main/playstore_reviews.csv')
data.head
```

```python
df = data.drop(columns=['package_name'], axis=1)
df.head
```

FORMAT THE DATA TYPE INTO A COMMON STRUCTURE (GET RID OF THE JUNK)

```python
df["review"] = df["review"].str.strip().str.lower()
print(df)
```

SPLIT THE TRAINING AND TEST DATA

```python
from sklearn.model_selection import train_test_split
X=df['review']
y=df['polarity']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=100)
```

TRANSFORM TEXT TO COUNT VECTORISER 

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer_model = CountVectorizer(stop_words='english')
X_train = vectorizer_model.fit_transform(X_train).toarray()
vectorizer_model.get_feature_names_out()
```

```python
len(vectorizer_model.get_feature_names_out())
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
X_test = vectorizer_model.transform(X_test)
y_pred = model.predict(X_test)
```

EVALUATING MODEL OUTCOMES USING METRICS - ACCURACY_SCORE AND PRECISION

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix: {conf_matrix}")

precision = precision_score(y_test, y_pred)
print(f"Precision: {precision}")

recall = recall_score(y_test, y_pred)
(f"Sensitivity: {recall}")
```

BERNOUlI MODEL

```python
from sklearn.naive_bayes import BernoulliNB
bnmodel = BernoulliNB()
bnmodel.fit(X_train, y_train)
```

GAUSSIAN MODEL

```python
from sklearn.naive_bayes import GaussianNB
gmodel = GaussianNB()
gmodel.fit(X_train, y_train)
```

RANDOM FOREST MODEL

```python
from sklearn.ensemble import RandomForestClassifier
rfclassifier = RandomForestClassifier(n_estimators =100)
rfclassifier.fit(X_train, y_train)
```

```python
rfy_pred = rfclassifier.predict(X_test)

rfaccuracy = accuracy_score(y_test, rfy_pred)
print(rfaccuracy)
```

```python
accuracy_scores = [0.7597765363128491, 0.770949720670391, 0.7486033519553073, 0.770949720670391]
models = ['Multinomial', 'Bernouli', 'Gaussian', 'Random forest']
df = pd.DataFrame(data=accuracy_scores, index=models, columns=['Accuracy'])
print(df)
```
